#### 11. [0.75 баллов] В каком количестве заказов есть товар, который стоит более 40% от суммы всего чека?

Возможно, будет удобно посчитать отдельно среднюю стоимость заказа, добавить ее в исходные данные и сделать необходимые проверки.

*Данный комментарий стоит воспринимать как подсказку к одному из вариантов решений задания. Если в вашем варианте решения он не нужнен, это не страшно*

```python
order_totals = pass_data.groupby('order_id')['price_number'].sum().reset_index()
order_totals = order_totals.rename(columns={'price_number': 'order_total'})

pass_data_with_totals = pass_data.merge(order_totals, on='order_id')

pass_data_with_totals['price_ratio'] = (pass_data_with_totals['price_number'] / pass_data_with_totals['order_total']) * 100

high_ratio_items = pass_data_with_totals[pass_data_with_totals['price_ratio'] > 40]

orders_with_high_ratio = high_ratio_items['order_id'].nunique()

print(f"Количество заказов, где есть товар стоимостью более 40% от суммы чека: {orders_with_high_ratio}")
```

#### 12. [0.75 баллов] Предположим, что в данных была ошибка и Diet Coke (choice_description), который стоил $1.25, должен был стоить 1.35. Скорректируйте данные в таблицы и посчитайте, на какой процент больше денег было заработано с этого товара. Не забывайте, что количество товара не всегда равно 1.

```python
original_total = pass_data['price_number'].sum()

diet_coke_mask = pass_data['choice_description'].str.contains('Diet Coke', na=False)
diet_coke_original = pass_data[diet_coke_mask].copy()

price_difference_per_unit = 1.35 - 1.25

pass_data.loc[diet_coke_mask, 'price_number'] += diet_coke_original['quantity'] * price_difference_per_unit

new_total = pass_data['price_number'].sum()

diet_coke_new = pass_data[diet_coke_mask]
diet_coke_original_total = diet_coke_original['price_number'].sum()
diet_coke_new_total = diet_coke_new['price_number'].sum()

diet_coke_revenue_increase = diet_coke_new_total - diet_coke_original_total
diet_coke_percentage_increase = (diet_coke_revenue_increase / diet_coke_original_total) * 100

total_revenue_increase = new_total - original_total
total_percentage_increase = (total_revenue_increase / original_total) * 100

print(f"Исходная общая выручка: ${original_total:.2f}")
print(f"Новая общая выручка: ${new_total:.2f}")
print(f"Общее увеличение выручки: ${total_revenue_increase:.2f}")
print(f"Общий процент увеличения: {total_percentage_increase:.4f}%")

print(f"Исходная выручка от Diet Coke: ${diet_coke_original_total:.2f}")
print(f"Новая выручка от Diet Coke: ${diet_coke_new_total:.2f}")
print(f"Увеличение выручки от Diet Coke: ${diet_coke_revenue_increase:.2f}")
print(f"Процент увеличения выручки от Diet Coke: {diet_coke_percentage_increase:.2f}%")
```

#### 3. [1 балл] Постройте средних сумм покупок по дням недели (bar plot).

```python
import datetime
import random

start = datetime.datetime(2018, 1, 1)
end = datetime.datetime(2018, 1, 31)
delta_seconds = int((end - start).total_seconds())

dates = pd.DataFrame(index=pass_data.order_id.unique())
dates['date'] = [
    (start + datetime.timedelta(seconds=random.randint(0, delta_seconds))).strftime('%Y-%m-%d')
    for _ in range(pass_data.order_id.nunique())]

pass_data['date'] = pass_data.order_id.map(dates['date'])
pass_data['date'] = pd.to_datetime(pass_data['date'])


pass_data['day_of_week'] = pass_data['date'].dt.dayofweek
daily_avg = pass_data.groupby('day_of_week')['price_number'].mean()

plt.figure(figsize=(10, 6))
daily_avg.plot(kind='bar', color='black')
plt.title('Средние суммы покупок по дням недели')
plt.xlabel('День недели (0=Понедельник, 6=Воскресенье)')
plt.ylabel('Средняя сумма ($)')
plt.show()
```
